#### Example Case Study: Classifying handwritten digits (0-9) from the MNIST dataset.
<center><img src='MNIST.png' border=1 height=200 width=400></center>

- The MNIST data set contains handwritten single digits from 0 to 9. 
- This data set can be easily accessed with Keras.
- The data set ha 60,000 Training and 10,000 Testing Images.

```python
# Import required Libraries

# Importing Tensorflow and Keras
import tensorflow as tf

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
# Import MNIST dataset
from tensorflow.keras.datasets import mnist
```

```python
# Load the MNIST dataset as Training and Testing data
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
X_train.shape, y_train.shape
```

```python
# Reading one Image of the MNIST X_train data
X_train[10]
```

```python
# Viewing the Image 10 of X_train
plt.imshow(X_train[10])
```

```python
# Reshape data to add channel dimension (1 for grayscale)
X_train = X_train.reshape((X_train.shape[0], 28, 28, 1))
X_test = X_test.reshape((X_test.shape[0], 28, 28, 1))

# Normalize the pixel values to range [0, 1]
X_train = X_train / 255.0
X_test = X_test / 255.0

# Convert class labels to one-hot encoding
y_train_cat = to_categorical(y_train, num_classes=10)  # Correctly one-hot encode y_train
y_test_cat = to_categorical(y_test, num_classes=10)    # Correctly one-hot encode y_test
```

```python
# Checking y_train data
y_train
```

```python
# Checking the Shape of the y_train
y_train.shape
```

```python
# Reading the 10th value of the y_train
y_train[10]
```

```python

# Create the CNN model
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(4, 4), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))  # Output layer for 10 classes
```

```python
# 3. Compiling the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train_cat, epochs=10, batch_size=64, validation_data=(X_test, y_test_cat))
```

```python
# Evaluation

from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(X_test, y_test_cat)
print(f'Test Accuracy: {test_accuracy:.4f}')
print(f'Test Loss: {test_loss:.4f}')

# Generate predictions for the test set
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # Get the predicted class indices

# Classification report
print(classification_report(y_test, y_pred_classes))
```
